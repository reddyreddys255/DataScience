# An example of "Web scraping" method applied on Wikipedia

```python
#our usual starter pack libraries to start a web scraping
import pandas as pd
import urllib.parse
import urllib.request
import bs4
#we difine our URL of wikipedia
url = 'https://en.wikipedia.org/wiki/Wikipedia'
#define a Header
DataBase=[('Company','Description')]
#Objects of our DataBade
Company=[]
Description=[]
#a liste of random companies to do the example 
liste=['HSBC','Nestlé','MICROSOFT']

for val in liste :
    Company=val
    #here we set a couple of (key,value) that will be used by the "urlencore" to be convert into apercent-encoded ASCII
    #txt string and then posted by the "request" method
    values = {'search' : val}
    data = urllib.parse.urlencode(values) #convert to bytes
    data = data.encode('ascii') #convert to "ascii" text format
    req = urllib.request.Request(url, data) #POST the request back to the web site
    with urllib.request.urlopen(req) as response:
    #after recovering the response from your "request" , we start reading it
       the_page = response.read()
    #then recovering the data from the reading step by using the "BeautifulSoup" library 
    page = bs4.BeautifulSoup(the_page, "html")
    text=[]

#     study the source code to locate the tags that contain the description text of the company , 
#     here it was the <p> from a parent tag <div> whom have the class name "'mw-parser-output'"
    for item in page.find('div', {'class' : 'mw-parser-output'}).findAll({'p'}) :
        
        #print(item.getText())
        #collecte all the texts from this tags
        text.append(item.getText())
    
    #build the data base
    DataBase.append((Company,text))
    
#convert it into a dataframe
df=pd.DataFrame(DataBase)
#put the header on the right place
new_header = df.iloc[0]
df = df[1:] 
df.columns = new_header 



```
